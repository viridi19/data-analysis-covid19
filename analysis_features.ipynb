```
# Dataset from states
states = pd.read_csv('/content/files/states.csv')
states.head()
```

```
# Merge data from analysis
corona = covid.merge(states, left_on='state', right_on='State')
corona['potentials'] = corona['Population'] * 0.3
corona['potentials'] = corona['potentials'].astype(int)
corona['date'] = pd.to_datetime(corona['date'])
corona.head()
```

```
group_uf = corona.groupby('state')
uf = group_uf.tail(1).sort_values('cases', ascending=False).drop(columns=['date', 'region', 'Region', 'Area', 'Cities count', 'GDP', 'GDP rate', 'potentials']).set_index('state')
uf.style.background_gradient(cmap='Reds', subset=['cases','deaths'])
```
